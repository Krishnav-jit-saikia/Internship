# Question - 1

```python
from bs4 import BeautifulSoup
import requests
```

```python
page1 = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page1
```

```python
soup1 = BeautifulSoup(page1.content)
```

```python
titles = []

for i in soup1.find_all(class_ = "mw-headline"):
    titles.append(i.text)
    
titles
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Headers': titles})
```

```python
df1
```

# Question - 2

```python
page2 = requests.get("https://presidentofindia.nic.in/former-presidents")
page2
```

```python
soup2 = BeautifulSoup(page2.content)
```

```python
former_presidents = []
for i in soup2.find_all('div' , class_ = 'desc-sec'):
    former_presidents.append(i.text.replace('\n' , ' '))
    
former_presidents
```

```python
df2 = pd.DataFrame({"Former Presidents" : former_presidents})
df2
```

# Question -  5

```python
page4 = requests.get('https://www.cnbc.com/world/?region=world')
page4
```

```python
soup4 = BeautifulSoup(page4.content)
```

```python
Headline = []
Time = []

for i in soup4.find_all('a' , class_ = "LatestNews-headline"):
  Headline.append(i.text)

for i in soup4.find_all('time' , class_="LatestNews-timestamp"):
  Time.append(i.text)
```

```python
df5 = pd.DataFrame({'Headline': Headline , 'Time': Time})
df5
```

# Question - 6

```python
page6 = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page6
```

```python
soup6 = BeautifulSoup(page6.content)
```

```python
Paper_Title = []
Authors = []
Published_Data =[]

for i in soup6('h2', class_ = 'sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
  Paper_Title.append(i.text)

for i in soup6('span' , class_ = 'sc-1w3fpd7-0 dnCnAO'):
  Authors.append(i.text)

for i in soup6('span' , class_ = "sc-1thf9ly-2 dvggWt"):
  Published_Data.append(i.text)
```

```python
df6 = pd.DataFrame({'Paper_Title': Paper_Title , 'Authors': Authors , 'Published_Date' : Published_Data})
df6
```

# Question - 7

```python
page7 = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page7
```

```python
soup7 = BeautifulSoup(page7.content)
```

```python
Restaurant_name = []
Cuisine = []
Location = []
Ratings = []
Image_URL = []

for i in soup7.find_all(class_="restnt-name ellipsis"):
  Restaurant_name.append(i.text)

for i in soup7.find_all('div', class_="restnt-loc ellipsis"):
  Location.append(i.text)

for i in soup7.find_all('div', class_="restnt-rating rating-4"):
  Ratings.append(i.text)

for i in soup7.find_all('img', class_="no-img"):
  Image_URL.append(i['data-src'])
```

```python
df7 = pd.DataFrame({'Restaurant name': Restaurant_name , 'Location': Location , 'Ratings': Ratings , 'Image_URL': Image_URL})
df7
```
